# Speech Female

Turn female voice to male voice

```python
inputFile, fs, hfreq, hmag, mXres, xr = hprT.analysis(
    inputFile='../sms-tools-enhanced/sounds/speech-female.wav', window='hamming', 
    M=901, N=2048, t=-120, minSineDur=0.1, nH=150, minf0=120, maxf0=260, f0et=8, 
    harmDevSlope=0.5, outfolder = 'res_test2')
```

```python
# Parameters for transformation
freqs = 0.6
freqScaling = np.array([0,freqs,1,freqs])
freqStretching = np.array([0,1,1,1])
timeScaling = np.array([0, 0, 1,1])
timbrePreservation = 0.5
timbreScaling = 0.8
changeResEnv=1
timbreScalingRes=0.88      # Scale residual by this factor
maxFreqFrac=0.75
```

```python
hprT.transformation_synthesis(inputFile, fs, hfreq, hmag, mXres, xr, freqScaling = freqScaling, 
	freqStretching = freqStretching, timbrePreservation = timbrePreservation, 
	timeScaling = timeScaling, timbreScaling = timbreScaling, changeResEnv=changeResEnv, 
    timbreScalingRes=timbreScalingRes, maxFreqFrac=maxFreqFrac, outfolder = 'res_test2')
```
