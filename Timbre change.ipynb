## Speech Female to Male

Change female voice to male voice

```python
# Load and analyse sound file
inputFile, fs, hfreq, hmag, mYst, stocEnv_nodecimate = hpsT.analysis(
    inputFile='../sms-tools-enhanced/sounds/speech-female.wav', window='hamming',
    M=901, N=2048, t=-120, minSineDur=0.1, nH=150, minf0=120, maxf0=260, f0et=8,
    harmDevSlope=0.5, stocf=0.2, outfolder='test/')
```

```python
# Parameters for the transformation
freqs = 0.625
freqScaling = np.array([0,freqs,1,freqs])
freqStretching = np.array([0,1,1,1])
timeScaling = np.array([0, 0, 1,1])
timbrePreservation = 0.5                   # Change timbre
timbreScaling = 0.8
#changeStocEnv=1
timbrePreservationStoc = 0                 # Change stochastic envelope timbre
timbreScalingStoc=0.85
maxFreqFrac=0.25
```

```python
# Apply the transformation
hpsT.transformation_synthesis(inputFile, fs, hfreq, hmag, mYst, stocEnv_nodecimate, stocf = 0.2, freqScaling = freqScaling, 
	freqStretching = freqStretching, timbrePreservation = timbrePreservation, 
	timeScaling = timeScaling, timbreScaling = timbreScaling,
    timbrePreservationStoc = timbrePreservationStoc,
    timbreScalingStoc = timbreScalingStoc,
    maxFreqFrac = maxFreqFrac, outfolder = 'test/')
```
